```python
import os
import json

# Obtém o diretório atual onde o notebook está localizado
diretorio = os.getcwd()

# Inicializa o total de filmes da América do Sul
total_filmes_sul_america = 0

# Lista de códigos de países da América do Sul
america_do_sul_codigos = ['AR', 'BO', 'BR', 'CL', 'CO', 'EC', 'FK', 'GF', 'GY', 'PY', 'PE', 'SR', 'UY', 'VE']

# Lista para armazenar os filmes da América do Sul
filmes_sul_america = []

# Loop pelos arquivos no diretório
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.json'):
        # Caminho completo para o arquivo
        caminho_arquivo = os.path.join(diretorio, arquivo)
        
        # Lê o conteúdo do arquivo JSON
        with open(caminho_arquivo, 'r') as f:
            conteudo_json = json.load(f)
            for filme in conteudo_json:
                if any(country['iso_3166_1'] in america_do_sul_codigos for country in filme.get('production_countries', [])):
                    total_filmes_sul_america += 1
                    filmes_sul_america.append(filme)

# Imprime o total de filmes da América do Sul
print("Total de filmes da América do Sul:", total_filmes_sul_america)

# Ordena os filmes da América do Sul por popularidade
filmes_sul_america.sort(key=lambda x: x.get('vote_average', 0), reverse=True)

# Imprime os top 100 filmes da América do Sul
print("\nTop 100 filmes da América do Sul:")
for i, filme in enumerate(filmes_sul_america[:100]):
    print(f"{i + 1}. {filme['title']} ({filme['vote_average']}) - {', '.join([country['name'] for country in filme['production_countries']])}")

# Conta quantos filmes de cada país da América do Sul estão no top 100
country_counts = {}
for filme in filmes_sul_america[:100]:
    for country in filme.get('production_countries', []):
        if country['iso_3166_1'] in america_do_sul_codigos:
            country_counts[country['name']] = country_counts.get(country['name'], 0) + 1

# Imprime o top 10 filmes de cada país da América do Sul
for country_name in country_counts.keys():
    country_movies = [filme for filme in filmes_sul_america if country_name in [c['name'] for c in filme['production_countries']]]
    country_movies.sort(key=lambda x: x.get('vote_average', 0), reverse=True)
    print(f"\nTop 10 filmes de {country_name}:")
    for i, filme in enumerate(country_movies[:10]):
        print(f"{i + 1}. {filme['title']} ({filme['vote_average']})")
```

```python

# Apresenta o número de filmes por país no top 100 da América do Sul
print("\nNúmero de filmes por país no top 100 da América do Sul:")
for country_name, count in country_counts.items():
    print(f"{country_name}: {count}")
```

```python
# Obtém o diretório atual onde o notebook está localizado
diretorio = os.getcwd()

# Inicializa contadores
total_filmes = 0
filmes_com_overview = 0
filmes_com_budget = 0
filmes_com_revenue = 0
filmes_sul_america = 0
filmes_sul_america_com_overview = 0
filmes_sul_america_com_budget = 0
filmes_sul_america_com_revenue = 0

# Lista de códigos de países da América do Sul
america_do_sul_codigos = ['AR', 'BO', 'BR', 'CL', 'CO', 'EC', 'FK', 'GF', 'GY', 'PY', 'PE', 'SR', 'UY', 'VE']

# Lista para armazenar os filmes da América do Sul
filmes_sul_america = []

# Loop pelos arquivos no diretório
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.json'):
        # Caminho completo para o arquivo
        caminho_arquivo = os.path.join(diretorio, arquivo)
        
        # Lê o conteúdo do arquivo JSON
        with open(caminho_arquivo, 'r') as f:
            conteudo_json = json.load(f)
            for filme in conteudo_json:
                total_filmes += 1
                if filme.get('overview'):
                    filmes_com_overview += 1
                if filme.get('budget'):
                    filmes_com_budget += 1
                if filme.get('revenue'):
                    filmes_com_revenue += 1
                if any(country['iso_3166_1'] in america_do_sul_codigos for country in filme.get('production_countries', [])):
                    filmes_sul_america.append(filme)

# Calcula as porcentagens
porcentagem_filmes_com_overview = (filmes_com_overview / total_filmes) * 100
porcentagem_filmes_com_budget = (filmes_com_budget / total_filmes) * 100
porcentagem_filmes_com_revenue = (filmes_com_revenue / total_filmes) * 100
total_filmes_sul_america = len(filmes_sul_america)
porcentagem_filmes_sul_america = (total_filmes_sul_america / total_filmes) * 100
for filme in filmes_sul_america:
    if filme.get('overview'):
        filmes_sul_america_com_overview += 1
    if filme.get('budget'):
        filmes_sul_america_com_budget += 1
    if filme.get('revenue'):
        filmes_sul_america_com_revenue += 1

# Calcula as porcentagens para os filmes da América do Sul
porcentagem_filmes_sul_america_com_overview = (filmes_sul_america_com_overview / total_filmes_sul_america) * 100
porcentagem_filmes_sul_america_com_budget = (filmes_sul_america_com_budget / total_filmes_sul_america) * 100
porcentagem_filmes_sul_america_com_revenue = (filmes_sul_america_com_revenue / total_filmes_sul_america) * 100

# Imprime os resultados
print("Total de filmes:", total_filmes)
print("Filmes com overview:", filmes_com_overview, f"({porcentagem_filmes_com_overview:.2f}% do total)")
print("Filmes com budget:", filmes_com_budget, f"({porcentagem_filmes_com_budget:.2f}% do total)")
print("Filmes com revenue:", filmes_com_revenue, f"({porcentagem_filmes_com_revenue:.2f}% do total)")
print("Total de filmes da América do Sul:", total_filmes_sul_america, f"({porcentagem_filmes_sul_america:.2f}% do total)")
print("Filmes da América do Sul com overview:", filmes_sul_america_com_overview, f"({porcentagem_filmes_sul_america_com_overview:.2f}% do total da América do Sul)")
print("Filmes da América do Sul com budget:", filmes_sul_america_com_budget, f"({porcentagem_filmes_sul_america_com_budget:.2f}% do total da América do Sul)")
print("Filmes da América do Sul com revenue:", filmes_sul_america_com_revenue, f"({porcentagem_filmes_sul_america_com_revenue:.2f}% do total da América do Sul)")
```
